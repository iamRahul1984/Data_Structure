2. How does broadcasting work in numpy

    - Broadcasting lets numpy perform operation on arrays of different shapes.smaller array are automatically expended to match larger ones

3. What is pandas DataFrame?
    
      - A DataFrame is a table-like structure with rows and columns. it is used to store,analyze,and manipulate data easily

     - groupby() is used to split data into groups based on a column . you can then apply functions like sum,mean or count

20. what is the role of a pivot table in Pandas?
